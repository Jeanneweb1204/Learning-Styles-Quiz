```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ikramshah512/amazon-products-sales-dataset-42k-items-2025")

print("Path to dataset files:", path)
import pandas as pd
```

```python
# Assign a new variable with the CSV path
csv_path = f"{path}/amazon_products_sales_data_cleaned.csv"

# Read the file
df = pd.read_csv(csv_path)
df # No print required, a
```

# üç™ Part 6 ‚Äî GroupBy & Aggregation

18. Compute the maximum discount for each product category.

```python
df_max_discount=df.groupby("product_category")[["discount_percentage"]].max()
df_max_discount
```

19. Find the number of products in each sponsorship type.

```python
df_number_of_products=df.groupby("is_sponsored")[["product_title"]].count()
df_number_of_products
```

20. Find the category with the highest average number of reviews per product.

```python
df_highest_avg_number_reviews=df.groupby("product_category")[["total_reviews"]].mean().sort_values(by="total_reviews",ascending=False)
df_highest_avg_number_reviews
```
